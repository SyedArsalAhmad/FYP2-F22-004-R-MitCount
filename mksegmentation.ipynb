```python
list_of_main_folders  = ['A03','A04','A05','A07','A10','A11','A12','A14','A15','A17','A18']
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt  

#dictionar of folder aolng with images
folder_dic = {}
import os
for folder in list_of_main_folders:
    path = r"E:\\FYP\\data\\training_aperio\\"+ folder +"\\frames\\x40"
    dir_list = os.listdir(path)
    folder_dic[folder] = dir_list
    
```

```python
folder_dic['A05']
```

```python
for folder in folder_dic.keys():
    for imgs in folder_dic[folder]:
        
        image1 = cv2.imread(folder+'\\frames\\x40\\'+imgs)
        
        img = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
# ret, thresh1 = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        ret, thresh1 = cv2.threshold(img,50, 200, cv2.THRESH_BINARY)
        source = cv2.GaussianBlur(thresh1, (7, 7), 0)

        mk=skimage.segmentation.mark_boundaries(img, source, color=(1, 1, 0), outline_color=None, mode='outer', background_label=0)
        
        mpimg.imsave(r'E:\FYP\data\training_aperio\mkaperioseg\ ' +str(imgs) ,mk)
        
        
```
